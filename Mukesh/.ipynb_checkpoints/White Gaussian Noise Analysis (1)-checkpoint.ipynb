```python
from random import gauss
```

```python
from random import seed
```

```python
from pandas import Series
```

```python
from pandas.tools.plotting import autocorrelation_plot
```

```python
seed(1)
```

```python
series = [gauss(0.0, 1.0) for i in range(1000)]
```

```python
series = Series(series)
```

```python
series.describe()
```

```python
import matplotlib.pyplot as pt
```

```python
series.plot()
```

```python
autocorrelation_plot(series)
```

# As we have plotted a Gaussian ditribution satisfying three conditions:
# 1) Zero Mean
# 2) Finite Variance
# 3) Uncorrelated lag (Zero autocorrelation)

# The given data exhibits a Gaussian White Noise Process

# The Ljung Box Statistic:
# alpha: 0.05
# Null hypothesis: No auto correlation exists between samples (lags)
# Alternate hypothesis: Data samples exhibit serial correlation

```python
import statsmodels.api
```

```python
statsmodels.stats.diagnostic.acorr_ljungbox(series, lags=None, boxpierce=False)
```

# All p-values resulted in probability values of greater than 5%. Thus we can accept the null hypothesis of no auto correlation between lags. Therefore zero auto correlation is proved with Ljung box Statistics

# Thus we can say that at any instant(sample), no pattern is observable on the time series data, due to which predictability will not be satisfactory
